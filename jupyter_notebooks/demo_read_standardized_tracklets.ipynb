```python
# NEOs...
df_neo = pd.read_csv(filepath_NEO, nrows=999999)  
df_neo.columns = df_neo.columns.str.replace(' ','')
df_neo.head()
```

```python
# Non-NEOs...
print(filepath_NONNEO)
df_nonneo = pd.read_csv(filepath_NONNEO, nrows=10)  
df_nonneo.columns = df_nonneo.columns.str.replace(' ','')
df_nonneo.head()
```
